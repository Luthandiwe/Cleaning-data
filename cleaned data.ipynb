```python
# Import required libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Airbnb_Open_Data (1).csv')
df
```

```python
# STEP 4: Preview the data
print("Initial Dataset Shape:", df.shape)
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Print the column names of the DataFrame
print(df.columns)
```

```python
# Drop rows with missing values
df = df.dropna()
```

```python
# Fill missing 'country' and price' with "Unknown" before dropping rows with missing values
df['country'].fillna('Unknown', inplace=True)
df['price'].fillna('Unknown', inplace=True)
df.dropna(inplace=True)
df
```

```python
# Convert 'last review' to datetime
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
df
```

```python
# Replace NaT with a placeholder date
df['last review'].fillna(pd.Timestamp('2000-01-01'), inplace=True)

print("After Cleaning:", df.shape)
df
```

```python
df['minimum nights'] = df['minimum nights'].astype(int)
df
```

```python
df['number of reviews'] = df['number of reviews'].astype(int)
df
```

```python
df['reviews per month'] = df['reviews per month'].astype(float)
df
```

```python
df['calculated host listings count'] = df['calculated host listings count'].astype(int)
df
```

```python
df['review rate number'] = df['review rate number'].astype(int)
df
```

```python
df['availability 365'] = df['availability 365'].astype(int)
df
```

```python
# Convert 'last review' to datetime if it's not already
if not pd.api.types.is_datetime64_any_dtype(df['last review']):
    df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

# Now you can extract the year
df['last review'] = df['last review'].dt.year
df
```

```python
# Standardize column names to lowercase
df.columns = [col.lower() for col in df.columns]
df
```

```python
# Define path to save file in your Google Drive
save_path = '/content/drive/MyDrive/Airbnb_Open_Data (1).csv'
```

```python
# Save the cleaned dataset
df.to_csv(save_path, index=False)

print("Cleaned data saved to Google Drive at:", save_path)
df
```
